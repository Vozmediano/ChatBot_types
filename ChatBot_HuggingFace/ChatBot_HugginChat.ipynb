```python
# import argparse
import getpass #pedir contraseñas de forma segura
import os#para operaciones del sistema operativo
import traceback #para manejar trazas de error
from hugchat import hugchat
from hugchat.login import Login #para interactuar con el modelo de Huggingface

# Autentificacion en Hugging Face
email = input("Ingrese su correo electrónico: ")  # Solicita el correo electrónico
passwd = getpass.getpass("Ingrese su contraseña: ")  # Solicita la contraseña de manera segura

sign = Login(email, passwd)
cookies = sign.login()

# Guardar las cookies, se guardan en la ruta /cookies_snapshot
cookie_path_dir = "./cookies_snapshot"
sign.saveCookiesToDir(cookie_path_dir)

# Define la función para iniciar una conversación con el usuario
def start_conversation():
    # Crea una instancia de ChatBot con las cookies de autenticación
    chatbot = hugchat.ChatBot(cookies=cookies.get_dict())
    # Obtiene el ID de conversación actual y la inicia
    conversation_id = chatbot.new_conversation()
    # Mensaje de bienvenida
    print("ChatBot: Hola, soy un ChatBot. ¿En qué puedo ayudarte?")
    # Inicia un bucle de conversación infinito
    while True:
        # Lee el mensaje del usuario desde la entrada estándar
        user_message = input("Usuario: ")
        # Salir del bucle si el usuario ingresa "salir"
        if user_message.lower() == "salir":
            print("ChatBot: Hasta luego.")
            break
        # Cambia a la conversación actual
        chatbot.change_conversation(conversation_id)
        # Obtiene la respuesta del chatbot
        response = chatbot.query(user_message)
        # Imprime la respuesta del chatbot
        print("ChatBot:", response)
# Llama a la función para iniciar la conversación
start_conversation()
```
