```python
# Se importan AutoTokenizer y BlenderbotForConditionalGeneration de la librería transformers
# para manejar el modelo de lenguaje y el tokenizador.
# pipeline también se importa de transformers para crear los pipelines de traducción.

from transformers import AutoTokenizer, BlenderbotForConditionalGeneration, pipeline
```

```python
# Se crea un pipeline de traducción del inglés al español utilizando el modelo Helsinki-NLP/opus-mt-en-es.
# Se crea otro pipeline de traducción del español al inglés utilizando el modelo Helsinki-NLP/opus-mt-es-en.

traductor_ingles_espanol = pipeline("translation", model="Helsinki-NLP/opus-mt-en-es")
traductor_espanol_ingles = pipeline("translation", model="Helsinki-NLP/opus-mt-es-en")
```

```python
# Cargar el modelo y el tokenizador.
# mname es el nombre del modelo que se va a utilizar, en este caso "facebook/blenderbot-1B-distill"
# Hay mas modelos de BlenderBot como por ejemplo:
# facebook/blenderbot-400M-distill
# facebook/blenderbot-3B
# facebook/blenderbot-90M
# DialoGPT:
# microsoft/DialoGPT-small
# microsoft/DialoGPT-medium
# microsoft/DialoGPT-large

# La elección del modelo depende de la tarea específica que deseas realizar y del idioma en el que necesitas trabajar.
# Hugging Face tiene una amplia variedad de modelos preentrenados disponibles que cubren una gran gama de
# aplicaciones en procesamiento de lenguaje natural.

# Se carga el modelo BlenderbotForConditionalGeneration con from_pretrained.
# Se carga el tokenizador correspondiente con AutoTokenizer.from_pretrained.

mname = "facebook/blenderbot-1B-distill"
model = BlenderbotForConditionalGeneration.from_pretrained(mname)
tokenizer = AutoTokenizer.from_pretrained(mname)
```

```python
# Mensajes de despedida en varios idiomas

mensajes_despedida = [
    "goodbye", "salir", "byebye", "hasta luego", "adios", "chao", "ciao"
]
```

```python
# Funcion para generar respuestas.
# Se define la función generar_respuesta que toma como parámetro entrada_usuario, el mensaje del usuario en español.
# traductor_espanol_ingles traduce la entrada del usuario del español al inglés.
# tokenizer convierte el texto en inglés en tensores de entrada para el modelo.
# model.generate genera una respuesta en inglés utilizando el modelo.
# tokenizer.batch_decode decodifica los ids generados de vuelta a texto.
# traductor_ingles_espanol traduce la respuesta del inglés al español.
# La función devuelve la respuesta en español.

def generar_respuesta(entrada_usuario):
  entrada_ingles = traductor_espanol_ingles(entrada_usuario)[0]["translation_text"]
  inputs = tokenizer([entrada_ingles], return_tensors="pt")
  reply_ids = model.generate(**inputs)
  respuesta_espanol = traductor_ingles_espanol(tokenizer.batch_decode(reply_ids, skip_special_tokens=True))[0]["translation_text"]

  return respuesta_espanol
```

```python
print("Hola soy un bot y estoy aqui apra ayudarte")
```

```python
# Inicia un bucle infinito para la interacción con el usuario.
# Solicita al usuario ingresar un mensaje con input("Usuario: ").
# Convierte el mensaje del usuario a minúsculas y verifica si está en la lista mensajes_despedida.
# Si es así, imprime "Bot: Adios" y rompe el bucle, terminando la conversación.
# Si el mensaje del usuario no es una despedida, llama a la función generar_respuesta para obtener la respuesta del chatbot.
# Imprime la respuesta del chatbot.

while True:
  entrada_usuario = input("Usuario: ")
  if entrada_usuario.lower() in mensajes_despedida:
    print("Bot: Adios")
    break
  respuesta = generar_respuesta(entrada_usuario)
  print(f"Bot: {respuesta}")
```
