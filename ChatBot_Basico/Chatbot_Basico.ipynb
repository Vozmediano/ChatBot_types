```python
import json

# Función para cargar respuestas desde un archivo JSON
# Se define la función cargar_respuestas que toma como parámetro ruta_archivo, la ruta del archivo JSON que contiene las respuestas.
# Utiliza with open para abrir el archivo en modo lectura con codificación UTF-8.
# json.load(archivo) carga y devuelve el contenido del archivo JSON como un diccionario de Python.

def cargar_respuestas(ruta_archivo):
    with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        return json.load(archivo)

# Cargar las respuestas desde el archivo JSON
respuestas = cargar_respuestas('/content/Peliculas.json')

# Función para obtener una respuesta
# Se define la función obtener_respuesta que toma como parámetro entrada_usuario, el mensaje del usuario.
# Convierte entrada_usuario a minúsculas para evitar problemas con la capitalización.
# Recorre la lista de respuestas buscando coincidencias entre entrada_usuario y el nombre de una película (pelicula['nombre']).
# Si encuentra una coincidencia, devuelve el resumen de la película (pelicula['resumen']).
# Si no encuentra ninguna coincidencia, devuelve un mensaje de error.

def obtener_respuesta(entrada_usuario):
    entrada_usuario = entrada_usuario.lower()  # Convertir a minúsculas
    for pelicula in respuestas:
        if pelicula['nombre'].lower() == entrada_usuario:
            return pelicula['resumen']
    return "Lo siento, no entiendo tu pregunta."

# Función para interactuar con el chatbot
# Se define la función interactuar_con_chatbot que inicia la interacción con el usuario.
# Imprime un mensaje inicial para informar al usuario cómo terminar la conversación.
# Inicia un bucle infinito que solicita al usuario ingresar el título de una película.
# Si el usuario escribe "adiós", el bucle se rompe y se imprime un mensaje de despedida.
# Para otros mensajes, se llama a obtener_respuesta para buscar y mostrar la respuesta correspondiente.

def interactuar_con_chatbot():
    print("Escribe 'adiós' para terminar la conversación.")
    while True:
        entrada_usuario = input("Escribe solamente el titulo de tu pelicula: ")
        if entrada_usuario.lower() == 'adios':
            print("Chatbot: ¡Adiós!")
            break
        respuesta = obtener_respuesta(entrada_usuario)
        print(f"Chatbot: {respuesta}")

# Iniciar la interacción
interactuar_con_chatbot()
```
